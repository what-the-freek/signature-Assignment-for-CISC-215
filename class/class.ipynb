## Format Specifier

```python
txt = "for only {price:.3f} dollars"
print(txt.format(price = 49.513548))
```

```python
txt1 = "my name is {fname}, I'm {age}".format(fname = "john", age = 35)
print(txt1)

txt2 = "my name is {}, I'm {}".format("ada", 33)
print(txt2)
txt3 = "my name is {0}, I'm {1}".format("ada", 33)
print(txt3)
```
