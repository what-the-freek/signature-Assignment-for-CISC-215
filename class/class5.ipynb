```python
n = int(input())
sum = 0
for i in range(1, n-1):
    if n%i==0:
        sum = sum+i
if sum==n:
    print(f"{n} is a perfect number")
```

```python
x = int(input())
y = int(input())
print(x,y)
for i in range(2, y-1):
    is_prime = 0
    k = 2
    while k*k <= i:
        if i % k == 0:
            is_prime = 1
            break
        k += 1
    if is_prime==0:
        print(i)
```

```python
x = int(input())
y = int(input())
print(x,y)
for i in range(x, y+1):
    if i ==1:
        continue
    prime = 0
    for j in range(2, i-1):
        if (i*j==0):
            prime =1
            break
    if prime == 0:
        print(i)
```
