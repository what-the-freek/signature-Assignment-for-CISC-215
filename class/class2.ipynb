```python
x = [1,2]
y = [4,5]

for i in x:
    for j in y:
        print(i,j)
```

```python
x = [1,2]
y = [4,5]
i = 0
while i < len(x):
    j=0
    while j<len(y):
        print(x[i],y[j])
        j = j+1
    i = i+1
```

• Write a program to check if a string contains ‘!’ symbol

```python
y = 0
x=[44,232,56,2,10]
for c in x:
    if c%2==0:
        y=y+1
print(y)
```

```python
check = False
x = input()
for c in x:
    if c == "!":
        check = True
print(check)
```

```python
n = 5

for i in range(1, n+1):
    for j in range(1, i+1):
        print("/", end=" ")
    print()
```

```python
n = 5

for i in range(n, 0,-1):
    for j in range(i,0,-1):
        print("*", end=" ")
    print()
```
