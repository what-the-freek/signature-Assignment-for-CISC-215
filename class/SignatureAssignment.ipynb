### Signature Assignment

```python
def get_average(marks):
    sum = 0 
    for a in marks:
        sum = sum + int(a)
    avg = sum/len(marks)
    return avg

if __name__=="__main__":
    num = 3
    with open("grades.txt","w") as wf:
        for i in range(num):
            name = input("enter name")
            grades = input("enter marks")
            grades=grades.split(',')
            avg = get_average(grades)
            wf.write(f"{name}:{avg:.2f}\n")
    with open("grades.txt","r") as f:
        data = f.read()
        print(data)
```
