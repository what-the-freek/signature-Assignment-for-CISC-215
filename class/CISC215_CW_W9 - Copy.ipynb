Write a function find_max(a, b, c) that returns the largest of three numbers. Prompt the user for three integers and print the maximum value.

```python
def find_max(a,b,c):
    if a > b and a >c:
        print(a,"a") 
    elif b > a and b >c:
        print(b,"b")
    else:
        print(c,"c")

if __name__=="__main__":
    find_max(12,6,10)
    find_max(6,10,15)
    find_max(6,10,8)
    find_max(8,12,4)
```

Write a function read_file(filename) that tries to open and read a file. If the file does not exist, catch the FileNotFoundError and print a friendly message.

```python
def read_file(filename):
    try:
        with open(filename, "r") as f:
            data = f.read()
            print(data)
    except FileNotFoundError:
        print("Error: File Not Found")
    
if __name__=="__main__":
    filename = input()
    read_file(filename)
```

Create a file numbers.txt manually or in code that contains one number per line. Write a function average_from_file(filename) that reads all numbers and returns their average. Handle possible exceptions such as file not found or invalid data (non-numeric line).

```python
def avg_from_file(filename):
    try:
        with open(filename,"r") as f:
            sum = 0
            c = 0
            for line in f:
                c+=1
                sum = sum/c
            avg = sum +int(line)
        return avg, line
    except FileNotFoundError:
        print(f"{filename} not found")
    except ValueError:
        print(f"{filename} has incorrect value")

if __name__=="__main__":
    avg_from_file = avg_from_file("numbers.txt")
    print(avg_from_file)
        
```

Write a function get_total_sales(filename) that:

Reads a file containing daily sales (one per line).

Returns the total sales.

Handles FileNotFoundError and ValueError (if any line is not a valid number).

```python
def get_total_sales(filename):
    try:
        with open("sales.txt", "r") as f:
            c = 0
            for line in f:
                c = c + int(line)
        return c
    except FileNotFoundError:
        print(f"{filename} not found")
    except ValueError:
        print(f"{filename} has incorrect value")
        
if __name__=="__main__":
    get_total_sales =get_total_sales("sales.txt")
    if get_total_sales:
        print(get_total_sales)
```

1. Write a function get_average(marks) that returns the average of a list of marks.
2. Ask the user to enter names and marks for 3 students (marks separated by commas).
3. Store each studentâ€™s name and average in a file grades.txt.
4. Then read the file and display all student averages.

```python
def get_average(marks):
    sum = 0 
    for a in marks:
        sum = sum + int(a)
    avg = sum/len(marks)
    return avg

if __name__=="__main__":
    num = 3
    with open("grades.txt","w") as wf:
        for i in range(num):
            name = input("enter name")
            grades = input("enter marks")
            grades=grades.split(',')
            avg = get_average(grades)
            print(f"{name}:{avg:.2f}")
            wf.write(f"{name}:{avg:.2f}\n")
    with open("grades.txt","r") as f:
        data = f.read()
        print(data)
```
