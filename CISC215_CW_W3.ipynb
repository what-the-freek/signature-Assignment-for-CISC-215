Problem 1: Check if a number is positive and even.

Input: A number (e.g., 4)

Output: 'Positive and Even' or 'something else'


```python
x = input()
x = int(x)

if x >0 and x%2==0:
    print(x, "is even and positive") 
else:
    print(x, "is something else")
```

Problem 4: Check if a character is a vowel.

Input: Input: A character (e.g., 'a')

Output: Output: 'Vowel' or 'Not a vowel'


```python
x = input()
vowel = ["a","e","i","o","u","A","I","O","U","e","E"]
if x in vowel :
    print(x, "is a vowel")
else:
    print(x, "is not a vowel")
```

Problem 5: Check if a number is between 10 and 50, inclusive.

Input: Input: A number (e.g., 25)

Output: Output: 'Within range' or 'Out of range'


```python
x = input()
x = int(x)

if x >= 10 and x <= 50:
    print("in range")
else:
    print("out of range")
```

Problem 7: Check if a number is divisible by 4 or 6.

Input: Input: A number (e.g., 12)
